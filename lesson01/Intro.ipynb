# Введение

Меня зовут Анна Швецова, ко мне можно обращаться на ты, Аня, Анна, Аняня, хей. На вы тоже можно, но только если очень хочется, но тогда без отчества. Я по умолчанию тоже обращаюсь к вам на ты, при желании это можно изменить, сказав/написав мне о том, как к вам обращаться. 


Я больше люблю, когда мне пишут в телегу, а не на почту (контакты смотрите на странице предмета).

## Правила игры

* Нужно получить зачёт по каждой домашке (= сдать её и получить от меня ок).
* Корректность 40%, стиль 40%, тесты 20%. Корректность является блокирующей.
* Балл по домашке динамически меняется со временем
    * В первую неделю вы пожете получить до 10 баллов (коэффициент у всех получаемых вами баллов = 1)
    * Во вторую до 7 (коэффициент у всех получаемых вами баллов = 0.7)
    * В третью неделю и далее не больше 4 (коэффициент у всех получаемых вами баллов = 0.4)
* Итерации работают следующим образом: когда вы присылаете мне посылку, я проверяю её и ли бо пишу вам, что всё хорошо, либо выдаю пачку рекомендаций и говорю, до какого момента вам можно это исправить (если я ничего не сказала про вермя, считайте, что у вас есть ровно 7 дней с момента отправки моего ответа на вашу посылку).
* Если вы исправляете мои комментарии до указанного срока, у вас остаётся коэффициент, который был на момент сдачи первой интерации. Если вы не справляетесь, ваш коэффициент становится равен текущему и процесс продолжается. 
    * Пример: вы сдали домашку на первой неделе и отправили исправления позже указанного срока где-то в конце второй недели, а потом отправили вторую пачку исправлений на третьей недели, но в рамках указанного срока, после чего я ответила вам, что домашка сдана. Тогда вы получаете свои баллы с коэффициентом 0.7
* Все дедлайны не железобетонные, и если вы случайно относительно разово продолбались, можно договориться со мной   локальном разовом переносе лично для вас. Договариваться надо ДО дедлайна. Так вероятность, что я соглашусь, гораздо, гораздо выше. После тоже можно попробовать, но причина должна быть скорее уважительной, чем нет.
* Если я поставила вам слишком короткий дедлайн на исправления (так можно считать по разным причинам. Например, недооценила объем или так оно сильно накладывается на другие дедлайны), со мной тоже можно и нужно об этом поговорить.
* Если при итерациях вы не исправляете ошибки, которые я присылаю, и ничего про это не пишете (вопросы, возражения и т.д.) так, что мне приходится указывать на одно и то же несколько раз, я могу начать вас штрафовать за это.
